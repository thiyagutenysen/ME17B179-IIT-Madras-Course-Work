```python
import PIL
print(PIL.__version__)
```

```python
from PIL import Image
```

```python
# create an image object to work with
image = Image.open('29.ppm')
image.show()
```

```python
# change image format
image.save('retina.jpg')
```

```python
# change size of image
image.thumbnail((300,300))
image.show()
image.save('thumbnail.png')
```

```python
# rotate it by 90 degrees anti clockwise
image = Image.open('29.ppm')
image = image.rotate(90)
image.show()
image.save('90degree.ppm')
```

```python
# change color of image to black and white
image = Image.open('29.ppm')
image = image.convert(mode='L')
image.save('blacknwhite.ppm')
```

```python
# Gaussian Blur
from PIL import Image, ImageFilter
image = Image.open('29.ppm')
image = image.filter(ImageFilter.GaussianBlur())
image.save('normal_gaussian_blur.ppm')
```

```python
# Gaussian Blur 2
from PIL import Image, ImageFilter
image = Image.open('29.ppm')
image = image.filter(ImageFilter.GaussianBlur(15))
image.save('increased_gaussian_blur.ppm')
```

```python
# Open image within Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
image = Image.open('29.ppm')
plt.imshow(image)
```
