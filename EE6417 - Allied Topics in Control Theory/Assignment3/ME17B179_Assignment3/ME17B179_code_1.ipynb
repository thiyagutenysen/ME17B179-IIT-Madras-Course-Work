```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Degree centrality
def degree_cen(G):
  return nx.degree_centrality(G)
```

```python
# Eigenvector centrality
def eigenvector_cen(G):
  return nx.eigenvector_centrality(G)
```

```python
# katz centrality
def katz_cen(G):
  rho = max(nx.adjacency_spectrum(G))
  alpha = 1/(2*rho)
  return nx.katz_centrality(G, alpha)
```

```python
# page_rank centrality
def pagerank_cen(G):
  return nx.pagerank(G,0.5)
```

```python
# pretty print
def organize_print(d):
  for i in d.keys():
    print("Node", i, ':', np.array(str(d[i])))
```

```python
# Find Centralities
def print_all_cens(G):
  nx.draw(G, with_labels = True)
  plt.show()
  print(" Degree Centrality = ")
  organize_print(degree_cen(G))
  print(" EigenVector Centrality = ")
  organize_print(eigenvector_cen(G))
  print(" Katz Centrality = ")
  organize_print(katz_cen(G))
  print(" PageRank Centrality = ")
  organize_print(pagerank_cen(G))
```

```python
# Ring Graph with 5
G = nx.circulant_graph(5,range(1,2))
print_all_cens(G)
```

```python
# Star Graph with 5 nodes
G = nx.star_graph(5)
print_all_cens(G)
```

```python
# Line Graph with 5 nodes
G = nx.path_graph(5)
print_all_cens(G)
```
