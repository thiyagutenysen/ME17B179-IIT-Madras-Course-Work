```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pprint
import json
```

```
def plot_resnet_results(filename):

  df = pd.read_csv(filename)

  train_loss = df['Trainset Loss'].values
  train_acc = df['Trainset Accuracy'].values
  val_loss = df['Testset Loss'].values
  val_acc = df['Testset Accuracy'].values
  # train_top1_error = df['Trainset Top1 Error'].values
  # test_top1_error = df['Testset Top1 Error'].values
  # train_top5_error = df['Trainset Top5 Error'].values
  # test_top5_error = df['Testset Top5 Error'].values
  test_classwise_accuracy = df['Testset Classwise Accuracy'].iloc[-1]

  plt.style.use('seaborn')
  fig,(ax1,ax2) =  plt.subplots(1,2,figsize=(15,5))
  fig.suptitle(filename[:-4])

  ax1.plot(train_acc, label='accuracy')
  ax1.plot(val_acc, label = 'val_accuracy')
  ax1.set_xlabel('Epoch')
  ax1.set_ylabel('Accuracy')
  ax1.legend(loc='lower right')

  ax2.plot(train_loss, label='loss')
  ax2.plot(val_loss, label = 'val_loss')
  ax2.set_xlabel('Epoch')
  ax2.set_ylabel('loss')
  ax2.legend(loc='upper right')

  # ax3.plot(train_top1_error, label='Trainset Top-1 Error')
  # ax3.plot(test_top1_error, label = 'Valset Top-1 Error')
  # ax3.set_xlabel('Epoch')
  # ax3.set_ylabel('Top-1 Error')
  # ax3.legend(loc='upper right')

  # ax4.plot(train_top5_error, label='Trainset Top-5 Error')
  # ax4.plot(test_top5_error, label = 'Valset Top-5 Error')
  # ax4.set_xlabel('Epoch')
  # ax4.set_ylabel('Top-5 Error')
  # ax4.legend(loc='upper right')

  ax1.title.set_text('Accuracy vs Epochs Plot')
  ax2.title.set_text('Loss vs Epochs Plot')
  # ax3.title.set_text('Top 1 Error in Dataset vs Epochs Plot')
  # ax4.title.set_text('Top 5 Error in Dataset vs Epochs Plot')

  plt.show()
  print()
  print('Test Set Classwise Accuracy after Training')
  json_acceptable_string = test_classwise_accuracy.replace("'", "\"")
  d = json.loads(json_acceptable_string)
  pprint.pprint(d)
```

```
# results for cifar10 without cbam
plot_resnet_results('Results_NO_CBAM.csv')
```

```
# results for cifar10 with cbam
plot_resnet_results('Results_CBAM.csv')
```

```
# results for cifar100 with cbam
plot_resnet_results('Results_CBAM-1.csv')
```
