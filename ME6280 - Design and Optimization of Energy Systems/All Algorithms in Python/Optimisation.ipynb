```python
# One variable example

# Given
h = 9
Tinf = 300
e = 0.6
q = 900
sigma = 5.67*(10**-8)

#initialization
i=0
error = float('inf')
Tol = 1e-8
method = 1

#initial guess
T = 340

#iteration
while(error>Tol):
    if method==1:
        newT = ((q - e*sigma*(T**4 - Tinf**4))/h) + Tinf
    else:
        newT = ((q-h*(T-Tinf))/(e*sigma) + Tinf**4)**(1/4)
    error = (newT - T)**2
    i=i+1
    print('iteration =', str(i).ljust(4, ' '), 'T(i) =', str(T).ljust(19, ' '), 'T(i+1) =', str(newT).ljust(19, ' '), 'error =', error)
    T = newT
```

```python
# double variable

# given
a1 = 100
a2 = 9.5
a3 = 1.8
b1 = 20
b2 = -(4.8*(10**-5))
b3 = 2

#initialization
i=0
error = float('inf')
Tol = 1e-8

#initial guess
x1 = 120
x2 = 10

#iteration
while(error>Tol):
    X = np.array([x1, x2])
    f1 = x1 - a2*(x2**a3) - a1
    f2 = x2 - b2*(x1**b3) - b1
    F = np.array([f1, f2])
    J = np.array([[1, -(a2*a3*(x2**(a3-1)))], [-(b2*b3*(x1**(b3-1))), 1]])
    
    newX = X - (np.linalg.inv(J) @ F)
    
    new_x1 = newX[0]
    new_x2 = newX[1]
    error = ((new_x1-x1)/x1)**2 + ((new_x2-x2)/x2)**2
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_x1).ljust(19, ' '), 'x2(i) =', str(new_x2).ljust(19, ' '), 'error =', error)
    x1 = new_x1
    x2 = new_x2
```

```python
# Dichotomous Search

# Given
interval = [0.5,3.5]
Lo = interval[1]-interval[0]
delta = 0.001
max_eval = 8
max_iter = max_eval//2

def get_area(r):
    V = 4
    A = 2*np.pi*(r**2) + (2*V/r)
    return A

#iteration
for i in range(max_iter):
    L0 = interval[1]-interval[0]
    x1 = interval[1] - ((L0+delta)/2)
    x2 = interval[0] + ((L0+delta)/2)
    # two point test
    f1 = get_area(x1)
    f2 = get_area(x2)
    if f1<f2:
        new_interval = [interval[0],x2]
    elif f1>f2:
        new_interval = [x1,interval[1]]
    else:
        new_interval = [x1,x2]
    error = new_interval[1]-new_interval[0]
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_interval[0]).ljust(19, ' '), 'b(i) =', str(new_interval[1]).ljust(19, ' '), 'error =', str(error).ljust(19, ' '), 'RR =', Lo/error)
    interval = new_interval
```

```python
# Fibonacci Search

from functools import reduce
# Given
interval = [0.5,3.5]
Lo = interval[1]-interval[0]
max_eval = 8
max_iter = max_eval-1

def get_area(r):
    V = 4
    A = 2*np.pi*(r**2) + (2*V/r)
    return A

# fibonacci function
fibonacci = lambda n:reduce(lambda x,n:[x[1],x[0]+x[1]], range(n),[0,1])[0]

#iteration
for i in range(max_iter):
    I0 = interval[1]-interval[0]
    Fn = fibonacci(max_eval-i+1)
    Fn1 = fibonacci(max_eval-i)
    I1 = I0*(Fn1/Fn)
    print(I1)
    x1 = interval[1]-I1
    x2 = interval[0]+I1
    # two point test
    f1 = get_area(x1)
    f2 = get_area(x2)
    if f1<f2:
        new_interval = [interval[0],x2]
    elif f1>f2:
        new_interval = [x1,interval[1]]
    else:
        new_interval = [x1,x2]
    error = new_interval[1]-new_interval[0]
    if error==0:
        RR = 'inf'
        print("Fully Converged ")
    else:
        RR = Lo/error
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_interval[0]).ljust(19, ' '), 'b(i) =', str(new_interval[1]).ljust(19, ' '), 'error =', str(error).ljust(19, ' '), 'RR =', RR)
    if error==0: break;
    interval = new_interval
```

```python
# Golden Section Search

# Given
interval = [0.5,3.5]
Lo = interval[1]-interval[0]
max_iter = 8

def get_area(r):
    V = 4
    A = 2*np.pi*(r**2) + (2*V/r)
    return A

#iteration
for i in range(max_iter):
    I0 = interval[1]-interval[0]
    gold = ((5**0.5)-1)/2
    I1 = I0*gold
    x1 = interval[1]-I1
    x2 = interval[0]+I1
    # two point test
    f1 = get_area(x1)
    f2 = get_area(x2)
    if f1<f2:
        new_interval = [interval[0],x2]
    elif f1>f2:
        new_interval = [x1,interval[1]]
    else:
        new_interval = [x1,x2]
    error = new_interval[1]-new_interval[0]
    if error==0:
        RR = 'inf'
        print("Fully Converged ")
    else:
        RR = Lo/error
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_interval[0]).ljust(19, ' '), 'b(i) =', str(new_interval[1]).ljust(19, ' '), 'error =', str(error).ljust(19, ' '), 'RR =', RR)
    if error==0: break;
    interval = new_interval
```

```python
# Multi Variable Unidirectional Search

# Given

# def get_evaluation(x1, x2):
#     f = 8 + (x1**2)/2 + 2/(x1*x2) + 6*x2
#     return f

# initialization
X = np.array([1,1])
i=0
error = float('inf')
Tol = 1e-5

# iteration
while(error>Tol):
    new_x2 = (1/(3*X[0]))**0.5
    new_x1 = (2/new_x2)**(1/3)
    new_X = np.array([new_x1, new_x2])
    error = np.linalg.norm(new_X-X)
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_X[0]).ljust(19, ' '), 'x2(i) =', str(new_X[1]).ljust(19, ' '), 'error =', error)
    X = new_X
```

```python
# Normal Gradient Descent

# Given

# def get_evaluation(x1, x2):
#     f = 8 + (x1**2)/2 + 2/(x1*x2) + 6*x2
#     return f

# initialization
X = np.array([1,1])
alpha = 0.05
i=0
error = float('inf')
Tol = 1e-5
max_iter = 1000

# iteration
while(error>Tol and i<max_iter):
    
    grad_x1 = X[0]-(2/((X[0]**2)*X[1]))
    grad_x2 = (-2/(X[0]*(X[1]**2)))+6
    grad_f = np.array([grad_x1, grad_x2])
    new_X = X - alpha*grad_f
    
    error = np.linalg.norm(new_X-X)
    
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_X[0]).ljust(19, ' '), 'x2(i) =', str(new_X[1]).ljust(19, ' '), 'error =', error)
    X = new_X
```

```python
# Congugate Gradient Method

# initialization
X = np.array([0,0])
i=0
error = float('inf')
Tol = 1e-5
max_iter = 1000

def get_gradients(x1,x2):
    grad_x1 = 1 + 4*x1 + 2*x2
    grad_x2 = -1 + 2*x1 + 2*x2
    return np.array([grad_x1, grad_x2])

def get_hessian(x1, x2):
    A = np.array([[4,2],[2,2]])
    return A

# First step of conjugate gradient
grad_f = get_gradients(X[0], X[1])
S = -grad_f
A = get_hessian(X[0], X[1])
lamda = (grad_f[0]**2 + grad_f[1]**2)/(S.T @ A @ S)
new_X = X + lamda*S
error = np.linalg.norm(new_X-X)
print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_X[0]).ljust(19, ' '), 'x2(i) =', str(new_X[1]).ljust(19, ' '), 'error =', error)

# iteration
while(error>Tol and i<max_iter):

    grad_f = get_gradients(new_X[0], new_X[1])
    grad_f0 = get_gradients(X[0], X[1])
    A = get_hessian(new_X[0], new_X[1])
    
    #Conjugate direction
    beta = (np.linalg.norm(grad_f)/np.linalg.norm(grad_f0))**2;
    S = -grad_f + beta*S;
    if S.any()==0:
        print("Fully Converged")
        break
    lamda = (grad_f[0]**2 + grad_f[1]**2)/(S.T @ A @ S)
    
    X = new_X
    new_X = X + lamda*S
    error = np.linalg.norm(new_X-X)
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_X[0]).ljust(19, ' '), 'x2(i) =', str(new_X[1]).ljust(19, ' '), 'error =', error)
```

```python
# Constrained Optimization Penalty

# Given
interval_x1 = [0.5, 1]
interval_x2 = [1, 2]
p = 1000
max_iter = 1000

def get_area(r, h, p):
    V = 2*np.pi*(r**2) + 2*np.pi*r*h + p*(((np.pi*(r**2)*h)-4)**2)
    return V

gold = ((5**0.5)-1)/2
#iteration
for i in range(max_iter):
    I0_x1 = interval_x1[1]-interval_x1[0]
    I0_x2 = interval_x2[1]-interval_x2[0]
    I1_x1 = I0_x1*gold
    I1_x2 = I0_x2*gold
    r1 = interval_x1[1]-I1_x1
    r2 = interval_x1[0]+I1_x1
    h1 = interval_x2[1]-I1_x2
    h2 = interval_x2[0]+I1_x2
    r = 0.5*(r1+r2)
    h = 0.5*(h1+h2)
    # two point test for r
    f1 = get_area(r1, h, p)
    f2 = get_area(r2, h, p)
    if f1<f2:
        new_interval_r = [interval_x1[0],r2]
    elif f1>f2:
        new_interval_r = [r1,interval_x1[1]]
    else:
        new_interval_r = [r1,r2]
    # two point test for h
    f1 = get_area(r, h1, p)
    f2 = get_area(r, h2, p)
    if f1<f2:
        new_interval_h = [interval_x2[0],h2]
    elif f1>f2:
        new_interval_h = [h1,interval_x2[1]]
    else:
        new_interval_h = [h1,h2]
        
    interval_x1 = new_interval_r
    interval_x2 = new_interval_h
    
A = 2*np.pi*(r**2) + 2*np.pi*r*h
v = np.pi*(r**2)*h
print('After Iteration =', str(i+1).ljust(7, ' '), 'P =', str(p).ljust(7, ' '), 'x1 =', str(r).ljust(19, ' '), 'x2 =', str(h).ljust(19, ' '), 'A =', str(A).ljust(19, ' '), 'V =', str(v).ljust(19, ' '))
    
```

```python
# Gauss-Newton 

# Given
X = np.array([10,41,79,139,202,298])
Y = np.array([3.1,11.8,21.1,29.8,37.4,42.5])
N = len(X)

# function specifics
def get_evaluation(a, b, x):
    f = a*(1-(np.exp(-1*b*x)))
    return f
def get_grada(a, b, x):
    grad_a = 1-(np.exp(-1*b*x))
    return grad_a
def get_gradb(a, b, x):
    grad_b = a*x*np.exp(-1*b*x)
    return grad_b

# initialize
a = 40
b = 0.005
Tol = 1e-10
max_iter = 10
i = 0
error = float('inf')

# iterate
while(error>Tol and i<max_iter):
    z1 = get_grada(a, b, X)
    z2 = get_gradb(a, b, X)
    Z = np.array([z1,z2]).T
    ZT = Z.T
    
    Yhat = get_evaluation(a, b, X)
    D = Y-Yhat
    RHS = ZT @ D
    delta_A = np.linalg.inv(ZT @ Z) @ RHS
    
    new_a = a + delta_A[0]
    new_b = b + delta_A[1]
    error = (delta_A[0]/a)**2 + (delta_A[1]/b)**2
    i = i+1
    
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_a).ljust(19, ' '), 'b(i) =', str(new_b).ljust(19, ' '), 'error =', error)
    
    a = new_a
    b = new_b
    
print("Predictions: ")
print(Yhat)
```
