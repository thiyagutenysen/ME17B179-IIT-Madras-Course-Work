```python
# Dichotomous Search

# Given
interval = [1,10]
Lo = interval[1]-interval[0]
delta = 0.001
max_eval = 40
max_iter = max_eval//2

def get_evaluation(v):
    QP = -((2.6 + (7*(v**0.3)))/(1.8 + (0.01*(v**2.1))))
    return QP

# Tol
Tol = 1e-3
error = float('inf')
i = 0
#iteration
while(error>Tol):
    L0 = interval[1]-interval[0]
    x1 = interval[1] - ((L0+delta)/2)
    x2 = interval[0] + ((L0+delta)/2)
    # two point test
    f1 = get_evaluation(x1)
    f2 = get_evaluation(x2)
    if f1<f2:
        new_interval = [interval[0],x2]
    elif f1>f2:
        new_interval = [x1,interval[1]]
    else:
        new_interval = [x1,x2]
    v = (new_interval[1]+new_interval[0])*0.5
    qp = get_evaluation(v)
    error = new_interval[1]-new_interval[0]
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_interval[0]).ljust(19, ' '), 'b(i) =', str(new_interval[1]).ljust(19, ' '), 'v(i) =', str(v).ljust(19, ' '), 'qp(i) =', str(qp).ljust(19, ' '), 'error =', str(error).ljust(19, ' '), 'RR =', Lo/error)
    interval = new_interval
```

```python
# Fibonacci Search

from functools import reduce
# Given
interval = [1,10]
Lo = interval[1]-interval[0]
max_eval = 20
max_iter = max_eval-1

def get_area(v):
    QP = -((2.6 + (7*(v**0.3)))/(1.8 + (0.01*(v**2.1))))
    return QP

# fibonacci function
fibonacci = lambda n:reduce(lambda x,n:[x[1],x[0]+x[1]], range(n),[0,1])[0]

# Tol
Tol = 1e-3
error = float('inf')
i = 0
#iteration
while(error>Tol):
    I0 = interval[1]-interval[0]
    Fn = fibonacci(max_eval-i+1)
    Fn1 = fibonacci(max_eval-i)
    I1 = I0*(Fn1/Fn)
    #print(I1)
    x1 = interval[1]-I1
    x2 = interval[0]+I1
    # two point test
    f1 = get_area(x1)
    f2 = get_area(x2)
    if f1<f2:
        new_interval = [interval[0],x2]
    elif f1>f2:
        new_interval = [x1,interval[1]]
    else:
        new_interval = [x1,x2]
    error = new_interval[1]-new_interval[0]
    if error==0:
        RR = 'inf'
        print("Fully Converged ")
    else:
        RR = Lo/error
    v = (new_interval[1]+new_interval[0])*0.5
    qp = get_area(v)
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_interval[0]).ljust(19, ' '), 'b(i) =', str(new_interval[1]).ljust(19, ' '), 'v(i) =', str(v).ljust(19, ' '), 'q/p(i) =', str(qp).ljust(19, ' '), 'error =', str(error).ljust(19, ' '), 'RR =', RR)
    if error==0: break;
    interval = new_interval
```

```python
# Gauss-Newton 

# Given
X = np.array([1.4,1.3,2.7,2.1,3.1,3.4,4.4,4.6,5.7,5.8])
Y = np.array([232739.2,236731.8,212226.8,225390.2,193265.2,191154.2,151174.2,149267.2,99156.8,94750.8])
N = len(X)

# function specifics
def get_evaluation(a, b, c, x):
    f = a + (b*(X**c))
    return f
def get_grada(a, b, c, x):
    grad_a = np.array([1]*len(x))
    return grad_a
def get_gradb(a, b, c, x):
    grad_b = x**c
    return grad_b
def get_gradc(a, b, c, x):
    grad_c = b*(np.log(x)*(x**c))
    return grad_c

# initialize
a = 250*(10**3)
b = -5000
c = 2.5
Tol = 1e-10
max_iter = 10
i = 0
error = float('inf')

# iterate
while(error>Tol and i<max_iter):
    z1 = get_grada(a, b, c, X)
    z2 = get_gradb(a, b, c, X)
    z3 = get_gradc(a, b, c, X)
    Z = np.array([z1,z2,z3]).T
    ZT = Z.T
    
    Yhat = get_evaluation(a, b, c, X)
    D = Y-Yhat
    RHS = ZT @ D
    delta_A = np.linalg.inv(ZT @ Z) @ RHS
    
    new_a = a + delta_A[0]
    new_b = b + delta_A[1]
    new_c = c + delta_A[2]
    error = (delta_A[0]/a)**2 + (delta_A[1]/b)**2 + (delta_A[2]/c)**2
    i = i+1
    
    print('iteration =', str(i).ljust(4, ' '), 'a(i) =', str(new_a).ljust(19, ' '), 'b(i) =', str(new_b).ljust(19, ' '), 'n(i) =', str(new_c).ljust(18, ' '), 'error =', error)
    
    a = new_a
    b = new_b
    c = new_c
    
print("Predictions: ")
print(Yhat)
```

```python
# Congugate Gradient Method
actual_X = np.array([0.201813,0.549789,0.865762,1.288098,1.741791,2.222954,2.806504,3.389966,3.961491])
actual_Y = np.array([0.522508,1.294212,1.808682,2.355305,2.83119,3.262058,3.705788,4.104502,4.438907])
Xs = np.log(actual_X)
Ys = np.log(actual_Y)

# initialization
X = np.array([0,0])
i=0
error = float('inf')
Tol = 1e-5
max_iter = 1000

def get_gradients(x1,x2,Xs):
    grad_x1 = -2*np.sum(Ys-(x1)-(x2*Xs))
    grad_x2 = -2*np.sum(Ys-(x1*Xs)-(x2*(Xs**2)))
    return np.array([grad_x1, grad_x2])

def get_hessian(x1, x2, Xs):
    A = np.array([[2*len(Xs),2*np.sum(Xs)],[2*np.sum(Xs),2*np.sum(Xs**2)]])
    return A

# First step of conjugate gradient
grad_f = get_gradients(X[0], X[1], Xs)
S = -grad_f
A = get_hessian(X[0], X[1], Xs)
lamda = (grad_f[0]**2 + grad_f[1]**2)/(S.T @ A @ S)
new_X = X + lamda*S
error = np.linalg.norm(new_X-X)
print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_X[0]).ljust(19, ' '), 'x2(i) =', str(new_X[1]).ljust(19, ' '), 'error =', error)

# iteration
while(error>Tol and i<max_iter):

    grad_f = get_gradients(new_X[0], new_X[1], Xs)
    grad_f0 = get_gradients(X[0], X[1], Xs)
    A = get_hessian(new_X[0], new_X[1], Xs)
    
    #Conjugate direction
    beta = (np.linalg.norm(grad_f)/np.linalg.norm(grad_f0))**2;
    S = -grad_f + beta*S;
    if S.any()==0:
        print("Fully Converged")
        break
    lamda = (grad_f[0]**2 + grad_f[1]**2)/(S.T @ A @ S)
    
    X = new_X
    new_X = X + lamda*S
    error = np.linalg.norm(new_X-X)
    i = i+1
    print('iteration =', str(i).ljust(4, ' '), 'x1(i) =', str(new_X[0]).ljust(19, ' '), 'x2(i) =', str(new_X[1]).ljust(19, ' '), 'error =', error)
```
