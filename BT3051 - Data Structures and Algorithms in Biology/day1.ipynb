```python
def euler53(n):
    factorials = []
    for i in range(1,n+1):
        for j in range(1,i+1):
            numerator=1
            for k in range(i-j+1,i+1):
                numerator*=k
            denominator=fact(j)
            c=numerator/denominator
            if c>1000000:
                factorials.append(c)
    return factorials
```

```python
def euler9():
    lists=[]
    for a in range(1,1000):
        for b in range(1,1000):
            c=1000-a-b
            if c**2==(a**2)+(b**2):
                lists.append(str(a)+str(b)+str(c))
    return lists
```

```python
def improved_divisor(n):
    sum=0
    i=1
    while i<=math.sqrt(n):
        if n%i==0:
            if n/i==i:
                sum+=1
            else:
                sum+=2
        i=i+1
    return sum
            
```

```python
def euler11(n):
    trinum=0
    i=1
    while(i>=0):
        trinum+=i
        divisors=improved_divisor(trinum)
        if divisors>=n+1:
            return trinum
        i=i+1
```

```python
euler11(500)
```
